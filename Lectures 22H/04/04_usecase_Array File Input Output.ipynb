### File Input and Output with Arrays

1. Create two one dimensional arrays. One array, say a1, has the numbers from 1 to 5 and the other array, say a2,  has the numbers from 5 to 9. The task here is to remove all the numbers in array a1 that are present in array a2. Save the resulted array on the disk. Finally, load the saved array from the disk to see the result.

```python
import numpy as np
```

```python
#creating two one dimesional arrays
a1 = np.array([1,2,3,4,5])
a2 = np.array([5,6,7,8,9])

#to remove the elements in a1 that are present in a2
result = np.setdiff1d(a1,a2)

#saving the array on the disk
np.save('./data/result_array', result)

#loading the saved array from the disk to see the result
np.load('./data/result_array.npy')
```

2. Load 'Iris.csv' keeping the text intact. Extract a coulmn with sepal lengths then compute it's mean, median and standard deviation. Normalize sepal length. After normalization, calculate the mean, median and standard deviation again.


```python
input_file = './data/Iris.csv'
```

```python
# Lets check the data
!head ./data/Iris.csv
```

```python
# Load the data. Since, ndarrays are homogeneous we skip the header row and the Species column. Also, we skip the Id column as it just represents the row number
irises = np.loadtxt(fname=input_file, delimiter=',', skiprows=1, usecols=(1,2,3,4))
```

```python
irises[:10]
```

```python
print(irises.shape)
```

```python
# Sepal length will have the index 0 on axis 1 in the ndarray
sepal_length = irises[:,1]
sepal_length[:10]
```

```python
mu, med, sd = np.mean(sepal_length), np.median(sepal_length), np.std(sepal_length)
print(mu, med, sd)
```

```python
Smax, Smin = sepal_length.max(), sepal_length.min()
S = (sepal_length - Smin)/(Smax - Smin)
S[:10]
```

```python
mu, med, sd = np.mean(S), np.median(S), np.std(S)
print(mu, med, sd)
```
