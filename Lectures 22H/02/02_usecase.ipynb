**Data** **Structures**

1. I am going for grocery shopping to buy bread, eggs, milk and cheese. There are 4 items that I want to purchase for now. While I am on my way to grocery store, I remember that I need to buy rice too. So, I add rice and I have 5 items to purchase. Before I start buyings things, I want to sort them so that I remember all of the things that I need to buy. Now, I want to see the first item that I would buy, buy the item and then delete it. Once the bought item is deleted, I want to see the remaining items. 
What data structure would you use for this scenario and how?

```python
# This is my shopping list
shoplist = ['bread', 'eggs', 'milk', 'cheese']

print('I have', len(shoplist), 'items to purchase.')

print('These items are:', end=' ')
for item in shoplist:
    print(item, end=' ')

print('\nI also have to buy rice.')
shoplist.append('rice')
print('My shopping list is now', shoplist)

print('I will sort my list now')
shoplist.sort()
print('Sorted shopping list is', shoplist)

print('The first item I will buy is', shoplist[0]) #use of indexing
olditem = shoplist.pop(0)
print('I bought the', olditem)
print('My shopping list is now', shoplist)
```

**Matplotlib Integration**

1. The Iris Dataset contains four features (sepals length, sepals width, petals length, and petals width) of 50 samples of three species of Iris (Iris setosa, Iris virginica and Iris versicolor).
*   Load the data provided to you using csv reader as a list
*   Scatter plot the first two features(sepal length and sepal width) of Iris Dataset.



```python
import matplotlib.pyplot as plt 
import csv
```

```python
input_file = './data/Iris.csv'
```

```python
with open(input_file, 'r') as iris_data:
    # The next statement, will ignore the first row which is the header
    next(iris_data)
    irises = list(csv.reader(iris_data))
irises
```

```python
irises[:10]
```

```python
sepal_length = []
sepal_width = []

#Can we do the following using comprehension
for i in range(0, len(irises)):
  sepal_length.append(float(irises[i][1]))
  sepal_width.append(float(irises[i][2]))

```

```python
plt.scatter(sepal_length, sepal_width)
plt.title('Scatter plot of Sepal  width againts Sepal Length ',  fontsize=12)
plt.xlabel('Sepal Length (cm)', fontsize=10)
plt.ylabel('Sepal Width (cm)', fontsize=10)
```
