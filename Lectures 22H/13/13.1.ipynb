- *Figures and Subplots*
- Plots in matplotlib reside within a *Figure* object
- Instead of plotting/displaying a plot, we may choose the create the figure object first to alter and change it before displaying
- A new figure can be created with *plt.figure* instance method:
```python
fig = plt.figure()
```
- This will create an empty plot window
- the *figsize* argument to plt.figure will ensure that the figure has a certain size and aspect ratio. It accepts a tuple of integers defining the width, height in inches.

````python
fig = plt.figure(figsize=(4,4))
````

- Subplots can be added to a plot figure by calling the instance method to figure *fig.add_subplot()*
- It returns an *AxesSubplot* object, that allows us to directly interact with the subplot
- It accepts three separate integers as arguments describing the position of the subplot. 
- If the three integers are R, C, and P in order, the subplot will take the **Pth position** on a grid with **R rows** and **C columns**.
- When dealing with subplots, the top *plt* instance, always refers to the newest subplot that is created.

````python
fig = plt.figure(figsize=(6,6))
# working on subplot 1 at top left in a 2x2 plotting space
ax1 = fig.add_subplot(2, 2, 1)
````

- Alternative to defining and creating a subplot indivisually, an array of subplots can be initialized while creating the plot figure itself
- the instance method *plt.subplots()* can be used to do this. 
- It accepts the layout of the subplot as shape of a 2-D array
- *sharex* and *sharey* boolean argument can also be specified to indicate that the subplots should have the same x- and y-axis respectively. By default, matplotlib autoscales plot limits independently
- It returns the parent figure and an array containing the AxesSubplots
- This is convenient as the axes array can be easily indexed as a 2-D array

````python
fig, axes = plt.subplots(2,3)
fig.subplots_adjust(wspace=.5, hspace=.5)
axes.shape
````

- *Adjusting the spacing around subplots*
- By default matplotlib leaves certain amount of padding around the outside of the subplots and between subplots
- This spacing is relative to the height and width of the plot and changes dynamically
- The spacing can be made static using the *subplots_adjust* instance method of Figure objects
- *wspace* and *hspace* arguments control the percent of figure width and height

````python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
  for j in range(2):
    axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
````

- *Ticks, Labels, and Legends*
- The *pyplot* interface is designed for interactive use and consists of methods like *xlim*, *xticks*, and *xticklabels*
- These control the plot range, tick locations and tick labels, repectively
- These can be used in two ways
  - Called with no arguments returns the current parameter value (e.g., *plt.xlim()* returns the current x-axis plotting range)
  - Called with parameters sets the parameter value (e.g., *plt.xlim([0, 10])*, sets the x-axis range from 0 to 10)
- All such methods used through the *plt* instance act on the active or most recently created AxesSubplot.
- Each of them also corresponds to two methods on the subplot object itself; in the case of *xlim* these are *ax.get_xlim* and *ax.set_xlim*

- *Setting the title, axis labels, ticks, ticklabels*

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
````

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

# Changing the x-axis ticks
# set_xticks instructs matplotlib on where to place the ticks along the data range;
#    by default these locations will also be the labels
ticks = ax.set_xticks(([0, 500, 1000]))
# With set_xticlabels, we can set alternate label values
labels = ax.set_xticklabels(['one', 'two', 'three'],
                           rotation=0, fontsize='large')
````

- Simillarly *set_yticks* and *set_yticklabels* can be used to set the ticks and labels for the y-axis
- Additionally, we can also provide a dict of options specifying formatting of a plot

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

props = {'title': 'My first matplotlib plot',
        'xlabel': 'Stages',
        'xticks': [0, 250, 500, 750, 1000],
        'xticklabels': ['one', 'two', 'three', 'four', 'five']
        }
_ = ax.set(**props)
````

- *Adding legends*
- Legends are another critical element for identifying plot elements
- There are a couple of ways to add them
- The most convenient way is to pass the *label* argument when adding each component of a plot
- After adding all the components to a plot, the *ax.legend()* or *plt.legend()* instance methods can be called to automatically create and place the legends in the plot

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
# loc=beast, Automatically determines the position the place the legend desc.
ax.legend(loc='best') 
````

- *Annotations and Drawing on a Subplot*
- Custom plot annotations such as text, arrows or other shapes can be created to a plot using the *text*, *arrow* and *annotate* instance methods
  - *text* draws text at a given coordinates (x, y) on the plot with optional custom styling
  ```python 
ax.text(x, y, 'Hello World!', family='monospace', fontsize=10)```
- Annotations can draw both text and arrow arranged appropiately.
- The *ax.set_xlim()* and *ax.set_ylim()* are recommanded to be used when creating annotation on plots rather than using matplotlib default limits

````python
# plotting the closing NASDAQ index for AAPL with the monthly High and Low
start = dt.date(day=1, month=9, year=2018)
end = dt.date(day=30, month=9, year=2018)
# First get the close index for sept 2018
aapl = datareader.DataReader('aapl', 'yahoo', start, end)['Close']
````

````python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(1, 1, 1)
ax.plot(aapl.reset_index(drop=True), 'k--')
ax.set_xticklabels(['d-'+str(d.day) for d in aapl.index], rotation=30, fontsize='small')
ax.set_title('AAPL closing index: ' + str(aapl.index.min().date()) + ' to ' + str(aapl.index.max().date()))

# Index of High and Low
ax.annotate(s='Low', xy=(aapl.reset_index(drop=True).idxmin(), aapl[aapl.idxmin()] ),
            xytext=(aapl.reset_index(drop=True).idxmin(), aapl[aapl.idxmin()] + 0.75 ) ,
            arrowprops=dict(facecolor='red', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='left', verticalalignment='top', color='blue')

ax.annotate(s='High', xy=(aapl.reset_index(drop=True).idxmax() + 0.5, aapl[aapl.idxmax()]),
            xytext=(aapl.reset_index(drop=True).idxmax() + 1.5, aapl[aapl.idxmax()]) ,
            arrowprops=dict(facecolor='green', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='left', verticalalignment='top', color='blue')
ax.set_xlim(0 - 1, aapl.shape[0] + 1)
ax.set_ylim(aapl.min() - 1, aapl.max() + 1)
````

- Additional shapes such as Rectangle, Cirle and Polygon can also be created on a matplotlib.pyplot
- A full set of shapes can be found in *matplotlib.patches*
- To add a shape to a plot, we create the patch object *shp* and add it to a subplot by calling *ax.add_patch(shp)*

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgn = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgn)
````

- *Saving Plots to File*
- A active figure can be saved to file using *plt.savefig(path/filename)* instance method
- The file type is inferred from the file extension. 
- Key options that can be specified while saving a file as arguments
  - *dpi*: controls the dots per inch resolution
  - *bbox_inches* can trim whitespace around the actual figure
  -- e.g.:
  ```python
plt.savefig('./images/figpath.png', dpi=400, bbox_inches='tight')```
